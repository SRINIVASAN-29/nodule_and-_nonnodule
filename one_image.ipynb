```python
X=df.iloc[:,:36].values
y=df.iloc[:,-1].values
```

### splitting data using sk-learn lib

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=1/3,random_state=20)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression   # logistic regression (LR)
from sklearn.tree import DecisionTreeClassifier   # tree classifier(DTC)
from sklearn.ensemble import RandomForestClassifier   # random tree classifier(RFC)

def my_model(x_train,y_train):
     LR=LogisticRegression(random_state=0)
     LR.fit(x_train,y_train)
     
     # tree classifier
     DTC=DecisionTreeClassifier(criterion="gini",random_state=0)
     DTC.fit(x_train,y_train)
     
     # random forest
     RFC=RandomForestClassifier(n_estimators=10,criterion="entropy",random_state=0)
     RFC.fit(x_train,y_train)
     
     print('the accurracy of logistic regression',LR.score(x_train,y_train))
     print('the accurracy of Decision Tree classifier',DTC.score(x_train,y_train))
     print('the accurracy of random forest classifier',RFC.score(x_train,y_train))
     
     return LR , DTC, RFC
```

```python
model=my_model(x_train,y_train)
```

### confusion matrix - it is a specific table layout that allows visualization of the performance of an algorithm

```python
from sklearn.metrics import confusion_matrix
matrix=confusion_matrix(y_test,model[0].predict(x_test))
tp=matrix[0][0]   #true positive
tn=matrix[1][1]   # true negative
fp=matrix[1][0]   # false positive
fn=matrix[0][1]   # false negative

print('confusion matrix \n',matrix)
print('accuracy',(tp+tn)/(tp+tn+fp+fn))
```

# Report

```python
from sklearn.metrics import accuracy_score,classification_report
for i in range(len(model)):
     print('MODEL is:',i)
     print(accuracy_score(y_test,model[i].predict(x_test)))
     print(classification_report(y_test,model[i].predict(x_test)))
```
